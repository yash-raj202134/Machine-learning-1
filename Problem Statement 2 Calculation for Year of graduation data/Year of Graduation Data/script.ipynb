# Calculation for Year of graduation from the given data'.

```python
import pandas as pd
```

```python
# Load the Excel dataset
df = pd.read_excel("Final Lead Data.xlsx") 
```

```python
df.sample(3)
```

```python
df.shape
```

```python
# Remove duplicate entries based on the "Email" column
df.drop_duplicates(subset=('Email'),keep='first',inplace=True)
```

```python
df.shape
```

```python
df
```

```python
# Checking for any null values in columns
df.isnull().sum()
```

```python
# Creating a new column "Branch" based on conditions
df['Branch'] = df.apply(lambda row: row['Other Branch'] if row['Branch/ Specialisation'] == 'Other' else row['Branch/ Specialisation'], axis=1)
```

```python
df.shape
```

```python
# Drop columns "Branch/ Specialisation" and "Other Branch"
df.drop(['Branch/ Specialisation', 'Other Branch'], axis=1, inplace=True)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df['Academic Year'].value_counts()
```

```python
df['What is your current academic year?'].value_counts()
```

```python
# Mapping the Current acedemic yr
mapping = {'1st Year': 1, '2nd Year': 2,'3rd Year':3,'Final Year':4}

# Replace values in the column using the dictionary
df['What is your current academic year?'] = df['What is your current academic year?'].replace(mapping)
```

```python
df['What is your current academic year?'].value_counts()
```

```python
# Creating a new column "Merged Academic Year" using combine_first()
df['Merged Academic Year'] = df['Academic Year'].combine_first(df['What is your current academic year?'])
```

```python
df.isna().sum()
```

```python
df.dropna(subset=['Merged Academic Year'], how='all', inplace=True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Merged Academic Year'].value_counts()
```

```python
# Calculating year of graduation from the merged academic year data obtained from above processes.
import datetime
current_year = datetime.datetime.now().year
# Custom function to calculate graduation year
def calculate_graduation_year(years_of_study):
    return int(current_year + (4.0 - years_of_study +1))

df['Year of Graduation'] = df['Merged Academic Year'].apply(calculate_graduation_year)
```

```python
df
```

```python
output_data = df[['ID','First Name','Email','Gender','City','Created','Branch','Merged Academic Year','Year of Graduation']].copy()
```

```python
output_data.reset_index(drop=True,inplace=True)
```

```python
# Saving the output in seperate excel file 'Result'
output_data.to_excel('Result.xlsx')
```
